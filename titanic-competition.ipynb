```python
train_data = pd.read_csv("train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("test.csv")
test_data.head()
```

```python
women=train_data.loc[train_data.Sex=='female']['Survived']
survival_percentage_of_women=(women.sum()/len(women))*100
survival_percentage_of_women
```

```python
men=train_data.loc[train_data.Sex=='male']['Survived']
survival_percentage_of_men=(men.sum()/len(men))*100
survival_percentage_of_men
```

```python
#checking for missing values
train_data.isnull().sum()
```

```python
test_data.isnull().sum()
```

```python
train_data.fillna(0)
test_data.fillna(0)
```

```python
from sklearn.ensemble import RandomForestClassifier
req_col=['Pclass','Sex','SibSp','Parch']
X=pd.get_dummies(train_data[req_col])
y=train_data["Survived"]
test_x=pd.get_dummies(test_data[req_col])
model=RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X,y)
pred_y=model.predict(test_x)
finalmodel={'PassengerId':test_data.PassengerId,'Survived':pred_y}
result=pd.DataFrame(finalmodel)
result.to_csv('submission.csv',index=False)
```
